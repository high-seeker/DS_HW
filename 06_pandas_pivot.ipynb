```python
# загрузить датасет в pandas dataframe
df = pd.read_csv('hotel_bookings.csv')
# показать все колонки
pd.options.display.max_columns = None
# вывести первые 5 и последние 5 строк
df
```

## Сводная таблица

```python
# взаимосвязь между 
# постоянством клиентов (is_repeated_guest), 
# типом отеля (hotel) и отменой брони (is_canceled)
pvt = df.pivot_table(index='is_repeated_guest', 
                     columns='hotel',
                     values='is_canceled',
                     aggfunc='sum')
pvt
```

```python
# взаимосвязь между 
# постоянством клиентов (is_repeated_guest),
# количеством детей (children)
# типом отеля (hotel) и отменой брони (is_canceled)
pvt = df.pivot_table(index='is_repeated_guest', 
                     columns=['children', 'hotel'],
                     values='is_canceled',
                     aggfunc='sum')
pvt
```

```python
# взаимосвязь между 
# постоянством клиентов (is_repeated_guest),
# типом депозита (deposit_type)
# типом отеля (hotel) и отменой брони (is_canceled)
# + привести обобщенные значения по каждой группе
pvt = df.pivot_table(index=['is_repeated_guest', 'deposit_type'], 
                     columns=['hotel'],
                     values='is_canceled',
                     aggfunc='sum',
                     margins=True)
pvt
```

```python
# используя функцию cut разбить значения 
# в столбце lead_time на 6 бинов (интервалов)
lead_time_bins = pd.cut(df['lead_time'], 6)

# вывести результат
lead_time_bins
```

```python
# число вхождений в каждый бин (интервал)
pd.value_counts(lead_time_bins)
```

```python
# взаимосвязь между 
# постоянством клиентов (is_repeated_guest),
# lead_time_intervals
# типом отеля (hotel) и отменой брони (is_canceled)
pvt = df.pivot_table(index=['is_repeated_guest', lead_time_bins], 
                     columns=['hotel'],
                     values='is_canceled',
                     aggfunc='sum')
pvt
```
