## Подготовка данных

```python
# добавить новый столбец (=all_days), который является суммой stays_in_week_nights (=weekdays) и stays_in_weekend_nights (=weekends)
df['all_days'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
df
```

```python
# сгруппировать данные по lead_time (функция агрегации - среднее значение)
df_lead_time = df.groupby(by=["lead_time"]).aggregate("mean")
df_lead_time
# сохранить в новую переменную сгруппированные значения для:
# all_days
# weekdays
# weekends

df = df.rename(columns={'stays_in_week_nights': 'weekdays'})
df = df.rename(columns={'stays_in_weekend_nights': 'weekends'})

df_lead_time_3 = df[["lead_time","all_days","weekdays","weekends"]]
df_lead_time_3 = df_lead_time_3.groupby(by=["lead_time"]).aggregate("mean")
df_lead_time_3
```

```python
# сгруппировать данные по arrival_date_month (функция агрегации - сумма)
df_arrival_date_month = df.groupby(by=["arrival_date_month"]).aggregate("sum")
df_arrival_date_month

# сохранить в новую переменную сгруппированные значения для is_canceled (=month_canceled)
df_month_canceled = df[df.is_canceled == 1].groupby(by=["arrival_date_month"]).is_canceled.aggregate("sum")
df_month_canceled

# сохранить в новую переменную сгруппированные значения для is_canceled ТОЛЬКО для летных месяцев (=summer_canceled)
df_summer = df[((df.arrival_date_month == 'July') | (df.arrival_date_month == 'June') | (df.arrival_date_month == 'August')) & (df.is_canceled == 1)]
df_summer = df_summer.groupby(by=["arrival_date_month"]).is_canceled.aggregate("sum")
df_summer
```

## Круговые диаграммы (pie chart)

### `Pyplot`-стиль

**ВНИМАНИЕ**: выделить сектор с **Максимальным** процентом.

![](./images/09/pie_plt.png)

```python
m = max(df_month_canceled)
explode = [0.15 if j == m else 0 for j in df_month_canceled]

patches, texts, autotexts = plt.pie(x=df_month_canceled, 
                                    explode=explode, 
                                    labels=df_month_canceled.index, 
                                    autopct='%1.1f%%',
                                    shadow=True, 
                                    startangle=0,                                   
                                    radius=1.5,
                                    textprops={'fontsize': 10})

plt.legend(title="Month",
           loc='best', 
           bbox_to_anchor=(1.05, 1))
```

## Ствол-лист диаграммы (stem plot)

### `Pyplot`-стиль

**ВНИМАНИЕ**: "базовая линия" (`bottom`) - это медиана.

![](./images/09/stem_plt.png)

```python
plt.stem(df_month_canceled.index, df_month_canceled,
         label='median',
         linefmt="m-.",
         markerfmt="g*",
         basefmt='y--',
         bottom= df_month_canceled.median())

plt.xticks(rotation=45)

plt.legend(loc='best')

plt.title(label='Stem Plot', 
          loc='center')
```

### ОО-стиль

**ВНИМАНИЕ**: "базовая линия" (`bottom`) - это медиана.

![](./images/09/stem_oo.png)

```python
fig, ax = plt.subplots(figsize=(8, 3))

ax.stem(df_month_canceled.index, df_month_canceled,
        label='median',
        linefmt="m-.",
        markerfmt="g*",
        basefmt='y--',
        bottom=df_month_canceled.median())

ax.xaxis.set_tick_params(which='both', rotation=70)
ax.yaxis.set_tick_params(which='both', rotation=-30)

ax.set_title(label='Stem Plot', 
             loc='center')

ax.legend(loc='best')
```

## Ящик с усами (Box Plot)

### `Pyplot`-стиль

**ВНИМАНИЕ**:
- особый стиль для отображения выбросов (outliers)

![](./images/09/box_plt.png)

```python
red_star_outliers = dict(marker="X",
                         markerfacecolor='#ffd17f',
                         markeredgecolor='#ff2c2c',
                         markersize=10, 
                         alpha=0.7)

_ = plt.boxplot(df_lead_time_3.all_days, 
                notch=True, 
                vert=False,
                flierprops=red_star_outliers)

plt.title(label='Stays nights', 
          loc='center')
```

### ОО-стиль

![](./images/09/box_oo.png)

```python
dict_with_mult_data = {'All': df_lead_time_3.all_days, 
                       'Weekday': df_lead_time_3.weekdays,
                       'Weekend': df_lead_time_3.weekends}

fig, ax = plt.subplots(figsize=(3, 5))

ax.boxplot(dict_with_mult_data.values(), 
           flierprops=red_star_outliers)

ax.set_xticklabels(dict_with_mult_data.keys())
```
